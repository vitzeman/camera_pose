a ta by nakonec rekonstruovala celý obrázek.

Toto promítání lze popsat matematickou rovnicí

$$ \mathbf{b}=\mathbf{P}\mathbf{B}. $$

Kde **b** je 2D bod obrazu (pixel), **B** je 3D bod v prostoru a **P** je matice kamery. **P** je matice celikosti 3 x 4, která hraje klíčovou roli při mapování reálného objektu na obrazovou rovinu. Lze ji také rozepsat pomocí rovnice

$$ \mathbf{P}=\mathbf{K}\mathbf{[R|t]}, $$

kde matice **K** popisuje vnitřní chování kamery, tj. jak je obraz vytvořen. A matice **[R|t]** určuje polohu a natočení kamery v prostoru. U jedné kamery se druhá část může zdát zbytečná, ale je nezbytná v případě interakcí více kamer.

Nyní, když jste pochopili pojem matice kamery

a vzorkujeme některé body na paprsku. Paprsek je běžně parametrizován
rovnicí `r(t) = o + td`, kde `t` je parametr, `o` je
počátek a `d` je jednotkový směrový vektor, jak je znázorněno na **Obrázku 6**.

# Praktická ukázka
Po dokončené instalaci a stažení datasetu můžeme konečně začít s praktickou ukázkou.

Jako první si nadefinujeme několik pomocných funkcí.

```python
def zakodovat_pozici(x):
    """Zakoduje pozici do odpovidajici Fourierovy funkce.

    Args:
        x: Vstupni souradnice.

    Returns:
        Fourierovy funkce tensoru pozice.
    """
    pozice = [x]
    for i in range(DIMENZE_POS_ENCODE):
        for fn in [tf.sin, tf.cos]:
            pozice.append(fn(2.0 ** i * x))
    return tf.concat(pozice, axis=-1)


def ziskat_paprsky(vyska, sirka, ohnisko, pozice):
    """Vypocita pocatecni bod a smerovy vektor paprsku.

    Args:
        vyska: Vyska obrazku.
        sirka: Sirka obrazku.
        ohnisko: Ohniskova vzdalenost mezi obrazky a kamerou.
        pozice: Matice pozice kamery.

    Returns:
        Dvojice pocatecniho bodu a smeroveho vektoru paprsku.
    """
    # Vytvorit meshgrid pro paprsky.
    i, j = tf.meshgrid(
        tf.range(sirka, dtype=tf.float32),
        tf.range(vyska, dtype=tf.float32),
        indexing="xy",
    )

    # Normalizovat souradnice osy x.
    transformovane_i = (i - sirka * 0.5) / ohnisko

    # Normalizovat souradnice osy y.
    transformovane_j = (j - vyska * 0.5) / ohnisko

    # Vytvorit smerove jednotkove vektory.
    smery = tf.stack([transformovane_i, -transformovane_j, -tf.ones_like(i)], axis=-1)

    # Ziskat matici kamery.
    matice_kamery = pozice[:3, :3]
    vyska_sirka_ohnisko = pozice[:3, -1]

    # Ziskat pocatky a smery paprsku.
    transformovane_smery = smery[..., None, :]
    smery_kamery = transformovane_smery * matice_kamery
    smery_paprsku = tf.reduce_sum(smery_kamery, axis=-1)
    pocatky_paprsku = tf.broadcast_to(vyska_sirka_ohnisko, tf.shape(smery_paprsku))

    # Vratit pocatky a smery.
    return (pocatky_paprsku, smery_paprsku)


def vykreslit_ploche_paprsky(pocatky_paprsku, smery_paprsku, blizko, daleko, pocet_vzorku, nahodne=False):
    """Vykresli paprsky a zplošti je.

    Args:
        pocatky_paprsku: Pocatecni body paprsku.
        smery_paprsku: Smerove jednotkove vektory paprsku.
        blizko: Blizka hranice objemove sceny.
        daleko: Vzdalená hranice objemove sceny.
        pocet_vzorku: Pocet vzorkovacich bodu na paprsku.
        nahodne: Volba pro nahodne vzorkovani.

    Returns:
       Dvojice zploštěných paprsků a vzorkovacích bodů na každém paprsku.
    """
    # Vypocitat 3D dotazovaci body.
    # Rovnice: r(t) = o+td -> Vytvoreni "t" zde.
    t_hodnoty = tf.linspace(blizko, daleko, pocet_vzorku)
    if nahodne:
        # Vlozit uniformni sum do vzorkovaciho prostoru, aby bylo vzorkovani
        # plynule.
        tvar = list(pocatky_paprsku.shape[:-1]) + [pocet_vzorku]
        sum = tf.random.uniform(shape=tvar) * (daleko - blizko) / pocet_vzorku
        t_hodnoty = t_hodnoty + sum

    # Rovnice: r(t) = o + td -> Vytvoreni "r" zde.
    paprsky = pocatky_paprsku[..., None, :] + (
        smery_paprsku[..., None, :] * t_hodnoty[..., None]
    )
    paprsky_ploche = tf.reshape(paprsky, [-1, 3])
    paprsky_ploche = zakodovat_pozici(paprsky_ploche)
    return (paprsky_ploche, t_hodnoty)




def mapovat_fn(pozice):
    """Mapuje jednotlivé pozice na zploštěné paprsky a vzorkovací body.

    Args:
        pozice: Matice pozice kamery.

    Returns:
        Dvojice zploštěných paprsků a vzorkovacích bodů odpovídajících
        pozici kamery.
    """
    (pocatky_paprsku, smery_paprsku) = ziskat_paprsky(vyska=V, sirka=S, ohnisko=ohnisko, pozice=pozice)
    (paprsky_ploche, t_hodnoty) = vykreslit_ploche_paprsky(
        pocatky_paprsku=pocatky_paprsku,
        smery_paprsku=smery_paprsku,
        blizko=2.0,
        daleko=6.0,
        pocet_vzorku=POCET_VZORKU,
        nahodne=True,
    )
    return (paprsky_ploche, t_hodnoty)
```

Dále si stáhnutý dataset připravíme do požadovaného formátu.



```python
# Vytvorit trenovaci rozdeleni.
index_rozdel = int(pocet_obrazku * 0.8)

# Rozdelit obrazky na trenovaci a validacni.
trenovaci_obrazky = obrazky[:index_rozdel]
val_obrazky = obrazky[index_rozdel:]

# Rozdelit pozice na trenovaci a validacni.
trenovaci_pozice = pozice[:index_rozdel]
val_pozice = pozice[index_rozdel:]

# Vytvorit trenovaci pipeline.
trenovaci_img_ds = tf.data.Dataset.from_tensor_slices(trenovaci_obrazky)
trenovaci_pozice_ds = tf.data.Dataset.from_tensor_slices(trenovaci_pozice)
trenovaci_paprsky_ds = trenovaci_pozice_ds.map(mapovat_fn, num_parallel_calls=AUTO)
trenovaci_ds = tf.data.Dataset.zip((trenovaci_img_ds, trenovaci_paprsky_ds))
tren_ds = (
    trenovaci_ds.shuffle(VELIKOST_DAVKY)
    .batch(VELIKOST_DAVKY, drop_remainder=True, num_parallel_calls=AUTO)
    .prefetch(AUTO)
)

# Vytvorit validacni pipeline.
val_img_ds = tf.data.Dataset.from_tensor_slices(val_obrazky)
val_pozice_ds = tf.data.Dataset.from_tensor_slices(val_pozice)
val_paprsky_ds = val_pozice_ds.map(mapovat_fn, num_parallel_calls=AUTO)
validacni_ds = tf.data.Dataset.zip((val_img_ds, val_paprsky_ds))
val_ds = (
    validacni_ds.shuffle(VELIKOST_DAVKY)
    .batch(VELIKOST_DAVKY, drop_remainder=True, num_parallel_calls=AUTO)
    .prefetch(AUTO)
)
```

## NeRF model

Model je vícevrstvý perceptron (MLP) s ReLU jako nelinearitou.

Úryvek z článku:

>"Podporujeme multiview-konzistentní reprezentaci tím, že
omezujeme síť na předpověď objemové hustoty sigma jako
funkci pouze polohy `x`, zatímco umožňujeme, aby byla RGB barva `c`
předpovídána jako funkce jak polohy, tak směru pohledu. Pro
dosažení tohoto cíle MLP nejprve zpracovává vstupní 3D souřadnici `x`
s 8 plně propojenými vrstvami (pomocí ReLU aktivací a 256 kanálů
na vrstvu) a výstupem je sigma a 256-dimenzionální vektor příznaků.
Tento vektor příznaků je pak zřetězen se směrem pohledu paprsku kamery
a předán do jedné další plně propojené vrstvy (používající
ReLU aktivaci a 128 kanálů), která vytváří výstup RGB barvy závislé na pohledu."

Zde jsme se rozhodli pro minimální implementaci a použili jsme 64
Dense jednotek místo 256, jak je uvedeno v článku.

```python
def ziskat_nerf_model(pocet_vrstev, pocet_poz):
    """Generuje NeRF neuronovou sit.

    Args:
        pocet_vrstev: Pocet vrstev MLP.
        pocet_poz: Pocet dimenzi pozicniho kodovani.

    Returns:
        `tf.keras` model.
    """
    vstupy = keras.Input(shape=(pocet_poz, 2 * 3 * DIMENZE_POS_ENCODE + 3))
    x = vstupy
    for i in range(pocet_vrstev):
        x = layers.Dense(units=64, activation="relu")(x)
        if i % 4 == 0 and i > 0:
            # Vlozit rezidualni spojeni.
            x = layers.concatenate([x, vstupy], axis=-1)
    vystupy = layers.Dense(units=4)(x)
    return keras.Model(inputs=vstupy, outputs=vystupy)



def vykreslit_rgb_hloubku(model, paprsky_ploche, t_hodnoty, nahodne=True, trenink=True):
    """Generuje RGB obraz a hloubkovou mapu z predikce modelu.

    Args:
        model: MLP model, ktery je trenovan na predikci rgb a
            objemove hustoty objemove sceny.
        paprsky_ploche: Zploštěné paprsky, které slouží jako vstup do
            NeRF modelu.
        t_hodnoty: Vzorkovaci body pro paprsky.
        nahodne: Volba pro nahodne vzorkovani.
        trenink: Zda je model ve fazi treninku nebo testovani.

    Returns:
        Dvojice RGB obraz a hloubkova mapa.
    """
    # Ziskat predikce z nerf modelu a preformatovat je.
    if trenink:
        predikce = model(paprsky_ploche)
    else:
        predikce = model.predict(paprsky_ploche)
    predikce = tf.reshape(predikce, shape=(VELIKOST_DAVKY, V, S, POCET_VZORKU, 4))

    # Rozdelit predikce na rgb a sigma.
    rgb = tf.sigmoid(predikce[..., :-1])
    sigma_a = tf.nn.relu(predikce[..., -1])

    # Ziskat vzdalenost sousednich intervalu.
    delta = t_hodnoty[..., 1:] - t_hodnoty[..., :-1]
    # tvar delta = (pocet_vzorku)
    if nahodne:
        delta = tf.concat(
            [delta, tf.broadcast_to([1e10], shape=(VELIKOST_DAVKY, V, S, 1))], axis=-1
        )
        alpha = 1.0 - tf.exp(-sigma_a * delta)
    else:
        delta = tf.concat(
            [delta, tf.broadcast_to([1e10], shape=(VELIKOST_DAVKY, 1))], axis=-1
        )
        alpha = 1.0 - tf.exp(-sigma_a * delta[:, None, None, :])

    # Ziskat transmitanci.
    exp_term = 1.0 - alpha
    epsilon = 1e-10
    transmitance = tf.math.cumprod(exp_term + epsilon, axis=-1, exclusive=True)
    vahy = alpha * transmitance
    rgb = tf.reduce_sum(vahy[..., None] * rgb, axis=-2)

    if nahodne:
        hloubkova_mapa = tf.reduce_sum(vahy * t_hodnoty, axis=-1)
    else:
        hloubkova_mapa = tf.reduce_sum(vahy * t_hodnoty[:, None, None], axis=-1)
    return (rgb, hloubkova_mapa)
```

## Trénování

Krok trénování je implementován jako součást vlastní podtřídy `keras.Model`,
abychom mohli využít funkcionalitu `model.fit`.

```python
class NeRF(keras.Model):
    def __init__(self, nerf_model):
        super().__init__()
        self.nerf_model = nerf_model

    def compile(self, optimizer, loss_fn):
        super().compile()
        self.optimizer = optimizer
        self.loss_fn = loss_fn
        self.loss_tracker = keras.metrics.Mean(name="loss")
        self.psnr_metric = keras.metrics.Mean(name="psnr")

    def train_step(self, inputs):
        # Ziskat obrazky a paprsky.
        (obrazky, paprsky) = inputs
        (paprsky_ploche, t_hodnoty) = paprsky

        with tf.GradientTape() as tape:
            # Ziskat predikce z modelu.
            rgb, _ = vykreslit_rgb_hloubku(
                model=self.nerf_model, paprsky_ploche=paprsky_ploche, t_hodnoty=t_hodnoty, nahodne=True
            )
            ztrata = self.loss_fn(obrazky, rgb)

        # Ziskat trenovatelne promenne.
        trenovatelne_promenne = self.nerf_model.trainable_variables

        # Ziskat gradienty trenovatelnych promennych vzhledem ke ztrate.
        gradienty = tape.gradient(ztrata, trenovatelne_promenne)

        # Aplikovat gradienty a optimalizovat model.
        self.optimizer.apply_gradients(zip(gradienty, trenovatelne_promenne))

        # Ziskat PSNR rekonstruovanych obrazku a zdrojovych obrazku.
        psnr = tf.image.psnr(obrazky, rgb, max_val=1.0)

        # Vypocitat nase metriky.
        self.loss_tracker.update_state(ztrata)
        self.psnr_metric.update_state(psnr)
        return {"loss": self.loss_tracker.result(), "psnr": self.psnr_metric.result()}

    def test_step(self, inputs):
        # Ziskat obrazky a paprsky.
        (obrazky, paprsky) = inputs
        (paprsky_ploche, t_hodnoty) = paprsky

        # Ziskat predikce z modelu.
        rgb, _ = vykreslit_rgb_hloubku(
            model=self.nerf_model, paprsky_ploche=paprsky_ploche, t_hodnoty=t_hodnoty, nahodne=True
        )
        ztrata = self.loss_fn(obrazky, rgb)

        # Ziskat PSNR rekonstruovanych obrazku a zdrojovych obrazku.
        psnr = tf.image.psnr(obrazky, rgb, max_val=1.0)

        # Vypocitat nase metriky.
        self.loss_tracker.update_state(ztrata)
        self.psnr_metric.update_state(psnr)
        return {"loss": self.loss_tracker.result(), "psnr": self.psnr_metric.result()}

    @property
    def metrics(self):
        return [self.loss_tracker, self.psnr_metric]
```

## Callback pro monitorování tréninku

Tento callback se používá k vizualizaci NeRF renderování během procesu trénování.

```python
testovaci_obrazky, testovaci_paprsky = next(iter(tren_ds))
testovaci_paprsky_ploche, testovaci_t_hodnoty = testovaci_paprsky

seznam_ztrat = []


class TreninkovyMonitor(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        ztrata = logs["loss"]
        seznam_ztrat.append(ztrata)
        testovaci_rekonstrukce_obrazky, hloubkove_mapy = vykreslit_rgb_hloubku(
            model=self.model.nerf_model,
            paprsky_ploche=testovaci_paprsky_ploche,
            t_hodnoty=testovaci_t_hodnoty,
            nahodne=True,
            trenink=False,
        )

        # Vykreslit rgb, hloubku a graf ztraty.
        fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
        ax[0].imshow(keras.preprocessing.image.array_to_img(testovaci_rekonstrukce_obrazky[0]))
        ax[0].set_title(f"Predikovany obrazek: {epoch:03d}")

        ax[1].imshow(keras.preprocessing.image.array_to_img(hloubkove_mapy[0, ..., None]))
        ax[1].set_title(f"Hloubkova mapa: {epoch:03d}")

        ax[2].plot(seznam_ztrat)
        ax[2].set_xticks(np.arange(0, EPOCHY + 1, 5.0))
        ax[2].set_title(f"Graf ztraty: {epoch:03d}")

        fig.savefig(f"obrazky/{epoch:03d}.png")
        plt.show()
        plt.close()


```

## Samotné trénování
Nyní je vše připraveno k tréninku.

```python
pocet_poz = V * S * POCET_VZORKU
nerf_model = ziskat_nerf_model(pocet_vrstev=8, pocet_poz=pocet_poz)

model = NeRF(nerf_model)
model.compile(
    optimizer=keras.optimizers.Adam(), loss_fn=keras.losses.MeanSquaredError()
)

# Vytvorit adresar pro ukladani obrazku behem treninku.
if not os.path.exists("obrazky"):
    os.makedirs("obrazky")

model.fit(
    tren_ds,
    validation_data=val_ds,
    batch_size=VELIKOST_DAVKY,
    epochs=EPOCHY,
    callbacks=[TreninkovyMonitor()],
    steps_per_epoch=index_rozdel // VELIKOST_DAVKY,
)


def vytvorit_gif(cesta_k_obrazkum, nazev_gifu):
    nazvy_souboru = glob.glob(cesta_k_obrazkum)
    nazvy_souboru = sorted(nazvy_souboru)
    obrazky = []
    for nazev_souboru in tqdm(nazvy_souboru):
        obrazky.append(imageio.imread(nazev_souboru))
    kargs = {"duration": 0.25}
    imageio.mimsave(nazev_gifu, obrazky, "GIF", **kargs)


vytvorit_gif("obrazky/*.png", "trenink.gif")
```

V této části žádáme model o vytvoření nových pohledů na scénu.

trénovacích obrázků nemůže obsahovat každý úhel
scény. Trénovaný model může reprezentovat celou 3D scénu s
řídkou sadou trénovacích obrázků.

Zde poskytujeme modelu různé pozice a žádáme ho, aby nám dal
2D obraz odpovídající tomuto pohledu kamery. Pokud provedeme inferenci modelu
pro všechny 360stupňové pohledy, mělo by to poskytnout přehled
celé scenérie ze všech stran.
